```python
import pandas as pd
```

```python
# Load the processed review sentiment DataFrame
df_sentiment = pd.read_pickle("../data/processed/app_level_sentiment.pkl")
```

```python
# Load the main merged DataFrame
main_df = pd.read_pickle("../data/processed/merged_data_with_copies_and_revenue.pkl")
```

```python
# Merge sentiment scores into the main DataFrame
merged_df = main_df.merge(df_sentiment[["app_id", "avg_sentiment"]], on="app_id", how="left")
```

```python
merged_df['avg_sentiment'].isna().value_counts()
```

```python
# Create a binary indicator for presence of sentiment score
merged_df['has_sentiment'] = merged_df['avg_sentiment'].notna().astype(int)
```

```python
merged_df.to_pickle("../data/processed/final_merged_data.pkl")
```
